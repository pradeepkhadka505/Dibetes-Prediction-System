At above table, we have seen some missing value, at here we have already clean such zero value of each column. just above method has shown now each colunm have not zero value. let's sets some new input value in each column having zero min value by replacing. **Let's check number of zero value in the data sets**

```python
print('The number of 0 value in column Gulcose:', data[data['Glucose']==0].shape[0])
print('The number of 0 value in column BloodPressure:', data[data['BloodPressure']==0].shape[0])
print('The number of 0 value in column SkinThickness:', data[data['SkinThickness']==0].shape[0])
print('The number of 0 value in column Insulin:', data[data['Insulin']==0].shape[0])
print('The number of 0 value in column BMI:', data[data['BMI']==0].shape[0])
```

**Replacing zero value with its repective mean value & some where meadian**

```python

data['Glucose']=data['Glucose'].replace(0,data['Glucose'].mean())#normal distribution
print('The number of 0 value in column Gulcose:', data[data['Glucose']==0].shape[0])

data['BloodPressure']=data['BloodPressure'].replace(0, data['BloodPressure'].mean())#normal distribution
print('The number of 0 value in column BloodPressure:', data[data['BloodPressure']==0].shape[0])

data['SkinThickness']=data['SkinThickness'].replace(0,data['SkinThickness'].mean())#normal distribution
print('The number of 0 value in column SkinThickness:', data[data['SkinThickness']==0].shape[0])

data['Insulin']=data['Insulin'].replace(0,data['Insulin'].median())#skewed distribution
print('The number of 0 value in column Insulin:', data[data['Insulin']==0].shape[0])

data['BMI']=data['BMI'].replace(0,data['BMI'].median())#skewed distribution
print('The number of 0 value in column BMI:', data[data['BMI']==0].shape[0])
```

**Step 7: Split data frame into X & y**

```python
target_name = 'Outcome'
#Seperate object for target features
y = data[target_name]
#serperate object for input freatures
X = data.drop(target_name, axis =1)
```

```python
data.head()
```

```python
y.head()
```

```python
#Applying features scalling techinque
scaler = StandardScaler()
scaler.fit(X)
x = scaler.transform(X)
```

```python
#Train the test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

**Step 8: Build classification Algorithms**

**1. Logistic Regression**
>Logistic regression is a type of regression analysis in statistics used for prediction of outcome of a categorical dependent variable from a set of predictor or independent variables. In logistic regression the dependent variable is always binary. Logistic regression is mainly used to for prediction and also calculating the probability of success.

```python
#sklearn logistic regressing model which is pre build model 
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear', multi_class='ovr')
lr.fit(X_train, y_train)
```

```python
X_test.shape
```

```python
#logistic regression prediction un y-test
lr_predict = lr.predict(X_test)
```

```python
#confusion Matrix Logistic Regression
lr_cm = confusion_matrix(y_test, lr_predict)
print(lr_cm)
```

```python
#seaborns heatmap library fror confusion matrix 
sns.heatmap(confusion_matrix(y_test, lr_predict), annot=True, fmt="d")
```

```python
#TrainScore and trainScore of linear Regression (train=80% & test -20%)
from sklearn.metrics import accuracy_score
print("Train accuracy of Logistic Regression:", lr.score(X_train, y_train)*100)
print("Accuracy(Test) score of Logistic Regression:", lr.score(X_test, y_test) *100)
print("Accuracy(Test) score of Logistic Regression:", accuracy_score(y_test, lr_predict)*100) #Actual Predictions
```

**I have find actual prediction accuracy of logistic regress: 81.81818181818183**

```python
print("Classification Report of Logistic Regression: \n", classification_report(y_test, lr_predict,digits=4))
```
